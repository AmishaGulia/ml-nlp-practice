```python
import nltk

# Download tokenizer (run once)
nltk.download('punkt')
```

```python
text = "Large language models are basically a type of neural network that has been trained on enormous volumes of textual data."
```

```python
# Tokenize text into words
from nltk.tokenize import word_tokenize

words = word_tokenize(text)
print("Tokens:", words)
```

```python
# Generate n-grams
from nltk.util import ngrams

# Unigrams
unigrams = list(ngrams(words, 1))

# Bigrams
bigrams = list(ngrams(words, 2))

# Trigrams
trigrams = list(ngrams(words, 3))
```

```python
# Display results
print("\nUnigrams:")
print(unigrams)

print("\nBigrams:")
print(bigrams)

print("\nTrigrams:")
print(trigrams)
```
