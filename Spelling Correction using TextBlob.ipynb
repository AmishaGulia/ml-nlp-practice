```python
from textblob import TextBlob
```

```python
# Input sentence with spelling errors
text = " Knowledge distillation transfers knwledge from a large model to a smller one without loss of vlidity."
```

```python
# Create TextBlob object
blob = TextBlob(text)

# Correct spelling
corrected_text = blob.correct()
```

```python
# Display results
print("Original Text:")
print(text)

print("\nCorrected Text:")
print(corrected_text)
```
