```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
# Ensure required NLTK resources are downloaded
nltk.download('punkt')
```

```python


raw_text = """
Natural language processing (NLP) is a field of computer science, artificial intelligence, and computational linguistics 
concerned with the interactions between computers and human (natural) languages. It involves tasks such as language modeling, 
sentiment analysis, and machine translation.
"""
# Step 1: Tokenize the text into sentences
sentences = sent_tokenize(raw_text)
# Step 2: Tokenize the text into words
words = word_tokenize(raw_text)

print("Tokenized Sentences:")
print(sentences)
print("\nTokenized Words:")
print(words)




```

```python

# Read your file
with open("Word_emb.txt", "r", encoding="utf-8") as file:
    text = file.read()

# Sentence Tokenization
sentences = sent_tokenize(text)
print("Sentences:")
for s in sentences:
    print(s)

# Word Tokenization
words = word_tokenize(text)
print("\nWords:")
print(words)
```
