Understand convex vs non-convex functions
Visualize:
Local minima
Global minima
See how optimization behaves visually

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

Local vs Global Minima  


```python
x = np.linspace(-5, 5, 500)
y = x**4 - 3*x**3 + 2  # Non-convex function

plt.figure(figsize=(8,5))
plt.plot(x, y)

# Mark approximate minima
plt.scatter([0, 2.25], [2, -6], color='red')  

plt.title("Local vs Global Minima")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()

plt.show()
```

```python
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)

Z = X**2 - Y**2  # Saddle function

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z)

ax.set_title("Saddle Point: f(x,y) = x^2 - y^2")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```

```python
x = np.linspace(-5, 5, 200)
y = x**2

plt.figure(figsize=(6,4))
plt.plot(x, y)

plt.title("Convex Function (Single Global Minimum)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()

plt.show()
```

```python
x = np.linspace(-5, 5, 500)
y = np.sin(x) + 0.1*x

plt.figure(figsize=(8,5))
plt.plot(x, y)

plt.title("Non-Convex Function (Many Local Minima)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()

plt.show()
```

```python
def f(x):
    return np.sin(x) + 0.1*x

def grad(x):
    return np.cos(x) + 0.1

x = 2.5  # starting point
lr = 0.1

trajectory = []

for _ in range(30):
    trajectory.append(x)
    x = x - lr * grad(x)

# Plot
xs = np.linspace(-5, 5, 500)
ys = f(xs)

plt.figure(figsize=(8,5))
plt.plot(xs, ys)
plt.scatter(trajectory, [f(i) for i in trajectory], color='red')

plt.title("Gradient Descent Path")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Convex function
def f(x):
    return x**2

x = np.linspace(-5, 5, 500)
y = f(x)

# Global minimum (for convex, only one)
min_x = x[np.argmin(y)]
min_y = np.min(y)

plt.figure(figsize=(7,5))
plt.plot(x, y, label="f(x) = x^2")

# Mark global minimum
plt.scatter(min_x, min_y, marker='o', s=100, label="Global Minimum")

plt.title("Convex Function (Single Global Minimum)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Non-convex function
def f(x):
    return np.sin(x) + 0.1*x

x = np.linspace(-10, 10, 1000)
y = f(x)

# Detect local minima
local_min_x = []
local_min_y = []

for i in range(1, len(x)-1):
    if y[i] < y[i-1] and y[i] < y[i+1]:
        local_min_x.append(x[i])
        local_min_y.append(y[i])

# Global minimum
global_min_index = np.argmin(y)
global_min_x = x[global_min_index]
global_min_y = y[global_min_index]

plt.figure(figsize=(10,6))
plt.plot(x, y, label="f(x) = sin(x) + 0.1x")

# Plot local minima
plt.scatter(local_min_x, local_min_y, marker='o', s=70, label="Local Minima")

# Highlight global minimum
plt.scatter(global_min_x, global_min_y, marker='*', s=200, label="Global Minimum")

plt.title("Non-Convex Function (Local vs Global Minima)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()

plt.show()
```

```python
def f(x):
    return np.sin(x) + 0.1*x

x = np.linspace(-10, 10, 1000)
y = f(x)

local_min_x, local_min_y = [], []
local_max_x, local_max_y = [], []

# Detect local minima & maxima
for i in range(1, len(x)-1):
    if y[i] < y[i-1] and y[i] < y[i+1]:
        local_min_x.append(x[i])
        local_min_y.append(y[i])
    elif y[i] > y[i-1] and y[i] > y[i+1]:
        local_max_x.append(x[i])
        local_max_y.append(y[i])

# Global extrema
global_min_x = x[np.argmin(y)]
global_min_y = np.min(y)

global_max_x = x[np.argmax(y)]
global_max_y = np.max(y)

# Plot
plt.figure(figsize=(10,6))
plt.plot(x, y, label="f(x) = sin(x) + 0.1x")

plt.scatter(local_min_x, local_min_y, s=70, label="Local Minima")
plt.scatter(local_max_x, local_max_y, s=70, label="Local Maxima")

plt.scatter(global_min_x, global_min_y, s=200, marker='*', label="Global Minimum")
plt.scatter(global_max_x, global_max_y, s=200, marker='X', label="Global Maximum")

plt.title("Non-Convex Function: Minima & Maxima")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()

plt.show()
```

#A convex function has:

Only one minimum
That minimum is both:
Local minimum
Global minimum
#A non-convex function has:

Multiple valleys → local minima
One lowest valley → global minimum
Non-Convex Case
Gradient descent may:
Stop at local minima
Miss global minimum
Local minima are often good enough
Saddle points are more problematic than minima
