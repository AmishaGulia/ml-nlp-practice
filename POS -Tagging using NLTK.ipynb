```python
# Import libraries
import nltk

# Download required resources run once
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
# Input sentence
sentence = "Low-Rank Adaptation (LoRA) is a parameter-efficient fine-tuning technique used to adapt large pre-trained models for specific tasks with minimal computational and memory overhead. As models grow larger, full fine-tuning becomes expensive. LoRA addresses this by reducing the number of trainable parameters, making the process faster and more efficient."
```

```python
# Tokenize sentence into words
from nltk.tokenize import word_tokenize

words = word_tokenize(sentence)
print("Tokens:", words)
```

```python
# Perform POS tagging
pos_tags = nltk.pos_tag(words)

print("\nPart-of-Speech Tags:")
for word, tag in pos_tags:
    print(f"{word} --> {tag}")
```
