No external libraries.
It uses basic Python string operations to perform morphological parsing.

```python
#Step 1: Input words
words = ["unhappiness", "redoing", "preprocessing", "misunderstood"]
```

```python
# Step 2: Define predefined prefixes and suffixes
prefixes = ["un", "re", "pre", "mis", "dis"]
suffixes = ["ness", "ing", "ed", "ly", "er"]
```

```python
# Step 3: Function to split word into prefix, stem, suffix
def morphological_parse(word):
    prefix = ""
    suffix = ""
    stem = word

    # Check prefix
    for p in prefixes:
        if word.startswith(p):
            prefix = p
            stem = word[len(p):]
            break

    # Check suffix
    for s in suffixes:
        if stem.endswith(s):
            suffix = s
            stem = stem[:-len(s)]
            break

    return prefix, stem, suffix
```

```python
# Step 4: Apply parser to each word
print("Morphological Parsing:\n")

for word in words:
    p, s, suf = morphological_parse(word)
    print(f"Word: {word}")
    print(f"  Prefix: {p}")
    print(f"  Stem: {s}")
    print(f"  Suffix: {suf}")
    print()
```

```python
#using NLTK Stemmer
# Import required modules
import nltk
from nltk.stem import PorterStemmer

# Download resources (only once)
nltk.download('punkt')
```

```python
# Input words
words = ["unhappiness", "redoing", "preprocessing", "misunderstood"]
```

```python
#Initialize stemmer
ps = PorterStemmer()
```

```python
#  Simple prefix & suffix lists
prefixes = ["un", "re", "pre", "mis", "dis"]
suffixes = ["ness", "ing", "ed", "ly", "er"]

def morphological_parse(word):
    prefix = ""
    suffix = ""
    
    # Detect prefix
    for p in prefixes:
        if word.startswith(p):
            prefix = p
            word = word[len(p):]
            break

    # Detect suffix
    for s in suffixes:
        if word.endswith(s):
            suffix = s
            word = word[:-len(s)]
            break

    # Use NLTK stemmer for better root extraction
    stem = ps.stem(word)

    return prefix, stem, suffix
```

```python
# Apply parser
print("Advanced Morphological Parsing:\n")

for word in words:
    p, s, suf = morphological_parse(word)
    print(f"Word: {word}")
    print(f"  Prefix: {p}")
    print(f"  Stem (NLTK): {s}")
    print(f"  Suffix: {suf}")
    print()
```
