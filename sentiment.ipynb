```python
# Import required libraries
import nltk
import string
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
```

```python
# Download necessary datasets (run once)
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
```

```python
# Define positive and negative words
positive_words = {"good", "great", "happy", "love", "excellent", "amazing", "nice", "awesome","best"}
negative_words = {"bad", "sad", "hate", "terrible", "poor", "worst", "awful"}
```

```python
# Convert POS tags to WordNet format
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return 'a'  # adjective
    elif tag.startswith('V'):
        return 'v'  # verb
    elif tag.startswith('N'):
        return 'n'  # noun
    elif tag.startswith('R'):
        return 'r'  # adverb
    else:
        return 'n'
```

```python
# Sample input (you can change this)
text = "I love this product! It is absolutely amazing and works great."
text = text.lower()

print("Original Text:", text)
```

```python
# Remove punctuation
text = text.translate(str.maketrans('', '', string.punctuation))

print("Cleaned Text:", text)
```

```python
nltk.download('punkt_tab')
```

```python
# Tokenize the text
tokens = word_tokenize(text)

print("Tokens:", tokens)
```

```python
import nltk

nltk.download('averaged_perceptron_tagger_eng')
```

```python
# Perform POS tagging
pos_tags = pos_tag(tokens)

print("POS Tags:")
print(pos_tags)
```

```python
# Initialize lemmatizer
lemmatizer = WordNetLemmatizer()

# Apply lemmatization with POS tags
lemmatized_tokens = [
    lemmatizer.lemmatize(word, get_wordnet_pos(tag))
    for word, tag in pos_tags
]

print("Lemmatized Words:", lemmatized_tokens)
```

```python
# Count positive and negative words
pos_count = sum(1 for word in lemmatized_tokens if word in positive_words)
neg_count = sum(1 for word in lemmatized_tokens if word in negative_words)

print("Positive Score:", pos_count)
print("Negative Score:", neg_count)
```

```python
# Determine sentiment
if pos_count > neg_count:
    sentiment = "Positive 😊"
elif neg_count > pos_count:
    sentiment = "Negative 😡"
else:
    sentiment = "Neutral 😐"

print("Final Sentiment:", sentiment)
```

```python
user_input = input("Enter your own sentence: ").lower()


clean = user_input.translate(str.maketrans('', '', string.punctuation))
tokens = word_tokenize(clean)
pos_tags = pos_tag(tokens)

lemmatized = [
    WordNetLemmatizer().lemmatize(word, get_wordnet_pos(tag))
    for word, tag in pos_tags
]

pos_count = sum(1 for word in lemmatized if word in positive_words)
neg_count = sum(1 for word in lemmatized if word in negative_words)

if pos_count > neg_count:
    print("Positive 😊")
elif neg_count > pos_count:
    print("Negative 😡")
else:
    print("Neutral 😐")
```
