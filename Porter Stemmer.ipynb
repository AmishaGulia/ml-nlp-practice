```python
##
The algorithm removes suffixes to reduce words to their root form.
Output may not always be a valid English word (e.g., "happiness" → "happi")
Porter Stemmer for speed and simple preprocessing.
#lemmatization when accuracy and valid root words are important.
```

```python
import nltk
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
```

```python
ps = PorterStemmer()
```

```python
text = "The workers are running quickly and building happiness in nations"
words = word_tokenize(text)
```

```python
stemmed_words = []

for word in words:
    stemmed_words.append(ps.stem(word))

print("Original Words:")
print(words)

print("\nStemmed Words:")
print(stemmed_words)
```
